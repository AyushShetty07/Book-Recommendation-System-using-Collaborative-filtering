## POPULARITY BASED RECOMMENDAION SYSTEM

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num-ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df

```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num-ratings']>=250].sort_values('avg_rating',ascending=False).head(50)

```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Image-URL-M','num-ratings','avg_rating']]
```

```python
popular_df
```

## colaborative based

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
good_users = x[x].index

```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data



```

```python
recommend('1984')
```

```python
import pandas as pd
import pickle

# Load the CSV file
csv_file = 'Ratings.csv'  # Replace with your CSV file path
df = pd.read_csv(csv_file)

# Specify the index of the row you want to add (e.g., the first row)
row_to_add = df.iloc[0]  # Replace 0 with the index of the desired row

# Load the existing pickle file
with open('books.pkl', 'rb') as pkl_file:
    existing_data = pickle.load(pkl_file)

# Print the type and structure of existing_data for debugging
print("Type of existing_data:", type(existing_data))
print("Contents of existing_data:", existing_data)

# Check the structure of existing_data
if isinstance(existing_data, pd.DataFrame):
    print("Shape of existing_data:", existing_data.shape)
    print("Columns of existing_data:", existing_data.columns.tolist())
elif isinstance(existing_data, list):
    print("Length of existing_data list:", len(existing_data))
else:
    raise ValueError("Unexpected type for existing_data: {}".format(type(existing_data)))

# Create a new row DataFrame for concatenation
new_row_df = row_to_add.to_frame().T  # Convert Series to DataFrame

# Check if the new row has the same columns as the existing data
if isinstance(existing_data, pd.DataFrame):
    if not new_row_df.columns.equals(existing_data.columns):
        raise ValueError("Column mismatch: Existing data columns do not match new row columns.")

    # Concatenate the DataFrames
    new_data = pd.concat([existing_data, new_row_df], ignore_index=True)
elif isinstance(existing_data, list):
    # If existing_data is a list, append the new row as a dictionary
    new_data = existing_data + [row_to_add.to_dict()]

# Save the combined data to a new pickle file
with open('Book.pkl', 'wb') as new_pkl_file:
    pickle.dump(new_data, new_pkl_file)

print("Row added to the new pickle file successfully!")
```
